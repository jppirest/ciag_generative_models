 ### The main idea of this notebook is to implement a MLP and use it as an autoencoder. We will use the toy dataset of FashionMNIST.

<img src = '../images/information-bottleneck.png' >

Autoencoders have many applications, such as **data compression, feature learning, dimensionality reduction** etc.


For the first notebook, we will use a MLP as our network. For toy datasets, such as the MNIST-based, the low dimensionality of data (1x28x28) allow us to use simple architectures. However, more complex image datasets aren't easily modelled using a MLP.


Nevertheless, for the second part we will implement a CNN as an autoencoder, given the fact that convolutional-based networks are more suited for image tasks, and we will use it to denoise an image.

## Architecture

This network should not be new to any of you. We will build a simple multilayer perceptron, using the `nn.Linear` layers and the `nn.ReLU` activation function. The images in this dataset are in *greyscale*, and the pixels are lying in the interval [0,1].
The core idea of this autoencoder is to reconstruct the original image. Applying a `nn.Sigmoid`, we will have pixels lying in the mentioned interval.


<img src = '../images/nn_ae.png' width = "550">


The model works as follows: 

1. 784 $\rightarrow$ 128 : `self.encode1`
2. 128 $\rightarrow$ 64 : `self.encode2`
3. 64 $\rightarrow$ 32 : `self.encode3`
4. 32 $\rightarrow$ 16 : `self.bottleneck`
5. 16 $\rightarrow$ 32 : `self.decode3`
6. 32 $\rightarrow$ 64 : `self.decode2`
7. 64 $\rightarrow$ 128 : `self.decode1`
6. 128 $\rightarrow$ 784 : `self.output`



All the layers, except  the `self.output`, is followed by the `nn.ReLU`.

One good implementation practice is to split the encoder and the decoder block. We will not draw separate classes for these blocks, due the simplicity of the network. But is good to have these blocks separated, to easily access the encode and decoder parts of the layer.

See the symmetry on the network? **Deep Learning is an empirical field.** There is no particular constraint for the symmetry, but most the success case were built upon symmetry constraints.

Finally, we also define a `latent_space_regularization` function. This function simply access the weights of the bottleneck layer (`self.bottleneck`), sum its absolute values and multiplies it by a small constant `lambda_`. 


```python
class Autoencoder(nn.Module):
    def __init__(self, input_shape : tuple = (1,28,28)):
        super().__init__()

        input = 1
        for element in input_shape:
            input*=element


        self.activation_function = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

        self.encode1 = nn.Sequential(nn.Linear(in_features=input, out_features=128), self.activation_function)
        self.encode2 = nn.Sequential(nn.Linear(in_features=128, out_features=64), self.activation_function)
        self.encode3 = nn.Sequential(nn.Linear(in_features=64, out_features=32), self.activation_function)
        self.bottleneck = nn.Sequential(nn.Linear(in_features=32, out_features=16), self.activation_function)
        self.decode3 = nn.Sequential(nn.Linear(in_features=16, out_features=32), self.activation_function)
        self.decode2 = nn.Sequential(nn.Linear(in_features = 32, out_features = 64), self.activation_function)
        self.decode1 = nn.Sequential(nn.Linear(in_features = 64, out_features = 128), self.activation_function)
        self.output = nn.Sequential(nn.Linear(in_features=128, out_features=input), self.sigmoid)


        encoders = [self.encode1, self.encode2, self.encode3, self.bottleneck]
        decoders = [self.decode3, self.decode2, self.decode1, self.output]

        self.encoder_block = nn.Sequential(*encoders)
        self.decoder_block = nn.Sequential(*decoders)


    def forward(self, x):
        x = x.reshape(-1, 784)
        out = self.encoder_block(x)
        out = self.decoder_block(out)
        out = out.reshape(-1,1,28,28)

        return out


    def latent_space_regularization(self):
        lambda_ = 0.01
        l1_loss = 0
        for param in self.bottleneck.parameters():
            l1_loss += torch.sum(torch.abs(param))
        return lambda_ * l1_loss
```

### Instantiating the model

```python
model = Autoencoder().to(device)

n_params = sum([p.numel() for p in model.parameters()])
n_params
```

```python
batch_size = 32
train_dataloader = DataLoader(train_data, batch_size = batch_size)
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3)
criterion = nn.MSELoss()
n_epochs = 10
```

### Training procedure

Apart from the `nn.MSELoss()` criterion, we will also sum to the loss the absolute value of the weights of the bottleneck layer. We will do this regularization technique because it motivates sparsity on the 16-dim latent space.

When training a network to classify, we should pass the labels to the loss function. For autoencoders, given the fact that we want to reconstruct our data, we pass the original input to the loss.

$$\mathcal{L} = \sum (\text{original image} - \text{reconstructed image})^2 + \sum \text{bottleneck's weights}$$


`loss = criterion(prediction, X) + model.latent_space_regularization()`

```python
model.train()

for epoch in tqdm(range(n_epochs)):
    for batch_idx, (X,y) in enumerate(train_dataloader):
        X = X.to(device)
        optimizer.zero_grad()
        prediction = model(X)
        loss = criterion(prediction, X) + model.latent_space_regularization()
        if not (torch.isnan(loss) | torch.isinf(loss)):
            loss.backward()
            optimizer.step()
        if (batch_idx + 1) % 250 == 0 and (epoch + 1) % 5 == 0:
            text = f'Loss: {loss}, Batch:{batch_idx + 1}/{len(train_dataloader)}, Epoch: {epoch + 1}'
            print(text)
```

### Validating the model on the test dataset

```python
batch_size = 256
test_dataloader = DataLoader(test_data, batch_size = batch_size, shuffle = True, drop_last = True)

original = np.zeros(shape = (len(test_dataloader), 256, 1, 28, 28))
predictions = np.zeros(shape = (len(test_dataloader), 256, 1, 28, 28))
labels = np.zeros(shape = (len(test_dataloader), 256))

model.eval()
with torch.no_grad():
    for batch_idx, (X,y) in enumerate(test_dataloader):
        X = X.to(device)
        pred = model(X)
        pred = pred.cpu().numpy()
        predictions[batch_idx] = pred
        original[batch_idx] = X.cpu().numpy()
        labels[batch_idx] = y.cpu().numpy()
```

```python
figure = plt.figure(figsize = (20,4))
cols, rows = 10, 2

for i in range (1, cols + 1):

    sample_idx = torch.randint(len(test_dataloader), size = (1,)).item()
    batch_idx = torch.randint(batch_size, size = (1,)).item()
    generated_image = predictions[sample_idx, batch_idx, 0]
    original_image = original[sample_idx, batch_idx, 0]

    figure.add_subplot(rows, cols, i)
    plt.axis('off')
    plt.imshow(generated_image, cmap='gray')
    plt.title('Generated Image')

    figure.add_subplot(rows, cols, i + cols)
    plt.axis('off')
    plt.imshow(original_image, cmap = 'gray')
    plt.title('Ground Truth')
plt.show()
```

### Exploring the latent space

We implemented both the encoder and decoder blocks as *self objects*.

Therefore, we can access these blocks separately and easily explore them.

We will now explore the latent space. Thus, we need to encode our data. Remember that the output of the decoder is 16-dimensional, as we defined in our architecture. Then, our vectors after encoded, will have shape `Batch Size x 16`.


```python
encoder = model.encoder_block

encoded_data = np.zeros(shape = (len(test_dataloader), batch_size, 16))
label =  np.zeros(shape = (len(test_dataloader), batch_size))

encoder.eval()
with torch.no_grad():
    for batch_idx, (X,y) in enumerate(test_dataloader):
        X = X.reshape(-1, 784)
        X = X.to(device)
        encoded_data[batch_idx] = encoder(X).cpu().numpy()
        label[batch_idx] = y.cpu().numpy()
```

```python
encoded_data = encoded_data.reshape(-1,16)
label_data = label.reshape(-1)

encoded_data = (encoded_data - encoded_data.min())/(encoded_data.max() - encoded_data.min())
```

```python
plt.style.use('default')
fig, ax = plt.subplots()


mean_vector = np.zeros(shape = (9, 16))
for i in range(9):
    vectors = encoded_data[label_data == i]
    vector = np.mean(vectors, axis = 0)
    mean_vector[i] = vector

ax.imshow(mean_vector)

for i in range(9):
    ax.text(-0.7, i, labels_map[i], va='center', ha='right')

for i in range(16):
    ax.text(i, -0.8, str(i), va='center', ha='center')

ax.axis('off');
```

We can see that in marked positions, the vector differ from one another, but in the most part of the vector, they are all very similar.

We will plot a 2d representation of our data, in which the choosen dimensions are the ones in which they are most different.


```python
layout = [
    ["A", "B", "C"],
    ["A", "D", "E"],
    ['A', 'F', 'G'],
    ['A', 'H', 'I'],
    ['A', "J", "K"]
]

auxiliar_axs = [chr(i) for i in range(ord('B'),ord('K')+1)]

plt.style.use('ggplot')

fig, axd = plt.subplot_mosaic(layout, layout = 'constrained', figsize=(12,6), width_ratios = [0.6,0.2,0.2] )


col1 = 10
col2 = 2
x = encoded_data[:,col1]
y = encoded_data[:,col2]
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'purple', 'orange', 'lime']

for color in np.unique(label_data):
    mask = (label_data == color)
    axd['A'].scatter(x[mask], y[mask], c = colors[int(color)], label = labels_map.get(color), alpha = 0.7)

axd['A'].legend()

for idx, aux in enumerate(auxiliar_axs):
    label_idx = torch.where(test_data.targets == idx)[0]
    random_idx = torch.randint(len(label_idx), size = (1,)).item()
    sample_idx = label_idx[random_idx]
    image, label  = test_data[sample_idx]
    axd[aux].imshow(image[0], cmap = 'gray_r')
    axd[aux].axis('off')
    axd[aux].set_title(labels_map.get(label))

    
```

We will use the cosine similarity for this task:

## Autoencoder with 2-dimensional bottleneck

Through the PCA we were able to see that rougly 2-dim are capable of explaining the 16-dim latent space.

A straightforward approach is to lower the bottleneck dimension to 2, and see if the NN is capable of reconstructing the data.
That's what we will do.

The architecture is very similar. Nevertheless, we have 1 more encode layer, `self.encode4`. This is because shrinking from 32 straight to 2 can sacrifice expressivity.

```python
class Autoencoder2(nn.Module):
    def __init__(self, input_shape : tuple = (28,28)):
        super().__init__()

        input = 1
        for element in input_shape:
            input*=element


        self.activation_function = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

        self.encode1 = nn.Sequential(nn.Linear(in_features=input, out_features=128), self.activation_function)
        self.encode2 = nn.Sequential(nn.Linear(in_features=128, out_features=64), self.activation_function)
        self.encode3 = nn.Sequential(nn.Linear(in_features=64, out_features=32), self.activation_function)
        self.encode4 = nn.Sequential(nn.Linear(in_features=32, out_features=16), self.activation_function)
        self.bottleneck = nn.Sequential(nn.Linear(in_features=16, out_features=2), self.activation_function)
        self.decode4 = nn.Sequential(nn.Linear(in_features=2, out_features=16), self.activation_function)
        self.decode3 = nn.Sequential(nn.Linear(in_features=16, out_features=32), self.activation_function)
        self.decode2 = nn.Sequential(nn.Linear(in_features = 32, out_features = 64), self.activation_function)
        self.decode1 = nn.Sequential(nn.Linear(in_features = 64, out_features = 128), self.activation_function)
        self.output = nn.Sequential(nn.Linear(in_features=128, out_features=input), self.sigmoid)


        encoders = [self.encode1, self.encode2, self.encode3, self.encode4, self.bottleneck]
        decoders = [self.decode4, self.decode3, self.decode2, self.decode1, self.output]

        self.encoder_block = nn.Sequential(*encoders)
        self.decoder_block = nn.Sequential(*decoders)


    def forward(self, x):
        x = x.reshape(-1, 784)
        out = self.encoder_block(x)
        out = self.decoder_block(out)
        out = out.reshape(-1,1,28,28)

        return out
    
    def latent_space_regularization(self):
        
        lambda_ = 0.01
        l1_loss = 0
        for param in self.bottleneck.parameters():
            l1_loss += torch.sum(torch.abs(param))
        return lambda_ * l1_loss
```

```python
model2 = Autoencoder2()

model2 = model2.to(device)
n_params = sum([p.numel() for p in model2.parameters()])
n_params
```

```python
optimizer = torch.optim.Adam(model2.parameters(), lr = 1e-4)
criterion = nn.MSELoss()
n_epochs = 10
```

```python
model2.train()

for epoch in tqdm(range(n_epochs)):
    for batch_idx, (X,y) in enumerate(train_dataloader):
        X = X.to(device)
        optimizer.zero_grad()
        prediction = model2(X)
        loss = criterion(prediction, X) + model2.latent_space_regularization()
        if not (torch.isnan(loss) | torch.isinf(loss)):
            loss.backward()
            optimizer.step()
        if (batch_idx + 1) % 100 == 0 and (epoch + 1) % 5 == 0:
            text = f'Loss: {loss}, Batch:{batch_idx + 1}/{len(train_dataloader)}, Epoch: {epoch + 1}'
            print(text)
```

```python
test_dataloader = DataLoader(test_data, batch_size = 256, shuffle = True, drop_last = True)

original = np.zeros(shape = (len(test_dataloader), 256, 1, 28, 28))
predictions = np.zeros(shape = (len(test_dataloader), 256, 1, 28, 28))
labels = np.zeros(shape = (len(test_dataloader), 256))

model2.eval()
with torch.no_grad():
    for batch_idx, (X,y) in enumerate(test_dataloader):
        X = X.to(device)
        pred = model2(X)
        pred = pred.cpu().numpy()
        predictions[batch_idx] = pred
        original[batch_idx] = X.cpu().numpy()
        labels[batch_idx] = y.cpu().numpy()
```

```python
figure = plt.figure(figsize = (20,4))
cols, rows = 10, 2

for i in range (1, cols + 1):
    sample_idx = torch.randint(len(test_dataloader), size = (1,)).item()
    batch_idx = torch.randint(batch_size, size = (1,)).item()
    generated_image = predictions[sample_idx, batch_idx, 0]
    original_image = original[sample_idx, batch_idx, 0]

    figure.add_subplot(rows, cols, i)
    plt.axis('off')
    plt.imshow(generated_image, cmap='gray')
    plt.title('Reconstructed')

    figure.add_subplot(rows, cols, i + cols)
    plt.axis('off')
    plt.imshow(original_image, cmap = 'gray')
    plt.title('Ground Truth')
plt.show()
```
