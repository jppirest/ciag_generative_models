# Module 5 - VAEs


 ### The main idea of this notebook is to implement a Variational Autoencoder and explore its properties. We will use the toy dataset of MNIST.

<img src = '../images/vae.png' >

Variational Autoencoders (VAEs) are a type of generative model used in unsupervised learning tasks. They were introduced to address several limitations of traditional autoencoders, *particularly in generating new data samples*.

At their core, VAEs are a combination of two fundamental components: an encoder and a decoder. These components work together to learn a compact representation of the input data and then reconstruct it. However, unlike standard autoencoders, VAEs also **enforce** a specific structure that enables them to generate new data samples.



```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor
import torch.nn as nn
from torch.utils.data import DataLoader
from tqdm.notebook import tqdm

seed = 42

np.random.seed(seed)
torch.manual_seed(seed)

if torch.cuda.is_available():
    torch.cuda.manual_seed(seed)
    device = 'cuda'
else:
    device = 'cpu'
```

```python
train_data = datasets.MNIST(root='/pgeoprj/ciag2023/datasets/mnist', train=True, download=True, transform=ToTensor(),)
test_data = datasets.MNIST(root='/pgeoprj/ciag2023/datasets/mnist', train=False, download=True, transform=ToTensor(),)
```

## Architecture

The overall "shape" of the architecture will be similar to a MLP autoencoder, which we implemented earlier.

We will use `nn.Linear` layers to both encode and decode our data. The chosen activation function is, once more, `nn.ReLU`.

Differently than in a standard Autoencoder, we will need to define two aditional functions `self.sampling` and the `self.kl_loss`, so we can introduce the tratability in our latent space.

The `self.sampling` is responsible of sampling an $\epsilon$ from a standard normal deviation and applying a transformation to the mean $\mu$ and $\sigma$, given by the encoder.

This transformation is pretty straightforward:

$$ z = \mu + \sigma \cdot \epsilon $$

We will use the `self.sampling` function inside the `self.encoder` function. This is because the sampling take both $\mu$ and $\sigma$ as inputs.

The `self.kl_loss` simply calculates the *Kullback-Leibler divergence* between the standard normal and the $\mu$, $\sigma$ given by the model. Minimizing this loss motivates the $\mu$, $\sigma$ take the normal gaussian shape.

```python
class VAE(nn.Module):

    def __init__(self):
        super().__init__()

        self.activation = nn.ReLU()
        self.encode1 = nn.Sequential(nn.Linear(in_features = 784, out_features = 128), self.activation)
        self.encode2 = nn.Sequential(nn.Linear(in_features = 128, out_features = 64), self.activation)
        self.mu = nn.Linear(in_features = 64, out_features = 2)
        self.sigma = nn.Linear(in_features = 64, out_features = 2)
        self.decode2 = nn.Sequential(nn.Linear(in_features = 2, out_features = 64), self.activation)
        self.decode1 = nn.Sequential(nn.Linear(in_features = 64, out_features = 128), self.activation)
        self.output = nn.Sequential(nn.Linear(in_features = 128, out_features = 784), nn.Sigmoid())

        encoders = [self.encode1, self.encode2]

        decoders = [self.decode2, self.decode1, self.output]

        self.encoder_block = nn.Sequential(*encoders)
        self.decoder_block = nn.Sequential(*decoders)

    def forward(self, x):

        encoded_data = self.encoder(x)

        decoded_data = self.decoder(encoded_data)

        return decoded_data


    def encoder(self, x):

        x = x.reshape(-1, 784)

        out = self.encoder_block(x)

        self.z_mean = self.mu(out)
        self.z_log_sigma = self.sigma(out)
        z = self.sampling(z_mean = self.z_mean, z_log_sigma = self.z_log_sigma)

        return z

    def decoder(self, z):

        out = self.decoder_block(z)
        out = out.reshape(-1,1,28,28)

        return out

    def sampling(self, z_mean, z_log_sigma):

        self.epsilon = torch.normal(mean = 0, std = 0.1, size = (z_mean.shape[0], z_mean.shape[1])).to(device)

        return z_mean + torch.exp(z_log_sigma) * self.epsilon

    def kl_loss(self):

        kl = -1*(1 + self.z_log_sigma - torch.pow(self.z_mean, 2) - torch.exp(self.z_log_sigma))
        kl = torch.sum(kl, axis = -1)

        return kl
```

## Instantiating the model

```python
model = VAE().to(device)

n_params = sum([p.numel() for p in model.parameters()])
n_params
```

```python
batch_size = 32
train_dataloader = DataLoader(train_data, batch_size = batch_size)
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3)
criterion = nn.MSELoss(reduction = 'none')
n_epochs = 20
```

## Training procedure

We set the reduction term of the `nn.MSELoss` to `none`, because we want the loss of each batch element. For each one of this batch elements, we sum its equivalent KL loss term. Finally, we take the mean.

$$\mathcal{L} = \frac{1}{n}\sum\limits_{i=0}^n (\mathbf{X}_i - \tilde{\mathbf{X}_i})^2 + \text{KL}(\mathcal{N}_{0,1}, p(\tilde{\mathbf{X}_i}))$$

```python
model.train()

for epoch in tqdm(range(n_epochs)):
    for batch_idx, (X,y) in enumerate(train_dataloader):
        X = X.to(device)
        optimizer.zero_grad()
        prediction = model(X)
        loss = torch.mean(criterion(prediction, X).sum(axis = (1,2,3)) + 0.5*model.kl_loss())
        if not (torch.isnan(loss) | torch.isinf(loss)):
            loss.backward()
            optimizer.step()
        if (batch_idx + 1) % 500 == 0 and (epoch + 1) % 2 == 0:
            text = f'Reconstruction Loss: {(loss - model.kl_loss()).sum().mean()}, KL Loss: {model.kl_loss().sum().mean()}, Batch:{batch_idx + 1}/{len(train_dataloader)}, Epoch: {epoch + 1}'
            print(text)
```

```python
b_size = 32
test_dataloader = DataLoader(test_data, batch_size = b_size, shuffle = True, drop_last = True)

original = np.zeros(shape = (len(test_dataloader),  b_size, 1, 28, 28))
predictions = np.zeros(shape = (len(test_dataloader),  b_size, 1, 28, 28))
labels = np.zeros(shape = (len(test_dataloader),  b_size))
encoded_data = np.zeros(shape = (len(test_dataloader),  b_size, 2))

model.eval()
with torch.no_grad():
    for batch_idx, (X,y) in enumerate(test_dataloader):
        X = X.to(device)
        pred = model(X)
        pred = pred.cpu().numpy()
        encoded = model.encoder(X)
        encoded = encoded.cpu().numpy()
        predictions[batch_idx] = pred
        encoded_data[batch_idx] = encoded
        original[batch_idx] = X.cpu().numpy()
        
        labels[batch_idx] = y.cpu().numpy()

encoded_data = encoded_data.reshape(-1, 2)
label_data = labels.reshape(-1)
```

```python
figure = plt.figure(figsize = (20,4))
cols, rows = 10, 2

for i in range (1, cols + 1):

    sample_idx = torch.randint(len(test_dataloader), size = (1,)).item()
    batch_idx = torch.randint(batch_size, size = (1,)).item()
    generated_image = predictions[sample_idx, batch_idx, 0]
    original_image = original[sample_idx, batch_idx, 0]

    figure.add_subplot(rows, cols, i)
    plt.axis('off')
    plt.imshow(generated_image, cmap='gray')
    plt.title('Generated Image')

    figure.add_subplot(rows, cols, i + cols)
    plt.axis('off')
    plt.imshow(original_image, cmap = 'gray')
    plt.title('Ground Truth')
plt.show()
```

## Clustering the latent space

There are several unsupervised algorithms for clustering data, each with its own strengths and weaknesses. The figure below demonstrates the behavior of each algorithm and the label assignments for different clustering techniques:

<img src='../images/clustering.png' width = '1000' >

For our case, we have the ground truth label of our data. So, we can compare the original labels with the labels given by the clustering.


All of these clustering algorithms are available in `sklearn.cluster`. However, the method of calling each one varies. For instance, with `HDBSCAN`, specifying the number of clusters is not necessary, whereas for the `SpectralClustering` routine, we must indicate the number of clusters present or expected in our data.

```python
X = np.vstack([x,y])

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, random_state=0, n_init="auto").fit(X.T)

#kmeans.labels_
```

## Latent Space Interpolation


The latent space given by the traditional autoencoders are somewhat messy, but mainly, not continuous. Thus, "interpolating" through this vectorial space is tricky.

<img src="../images/interpolation.png" >


VAEs, on the other hand, due to their probabilistic approach, make the latent space much easier to deal with. In addition to its continuity, the overall Gaussian distribution brings robustness.

<img src="../images/glasses.png" >

```python
n = 30
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))
# We will sample n points within [-4, 4] standard deviations

grid_x = np.linspace(-2, 2, n)
grid_y = np.linspace(-2, 2, n)

decoder = model.decoder

plt.style.use('default')

with torch.no_grad():
    for i, yi in enumerate(grid_x):
        for j, xi in enumerate(grid_y):
            z_sample = torch.tensor([[xi, yi]]).to(torch.float32).to(device)
            x_decoded = decoder(z_sample)
            digit = x_decoded[0].cpu().numpy()
            figure[i * digit_size: (i + 1) * digit_size,
                   j * digit_size: (j + 1) * digit_size] = digit

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(figure, cmap = 'gray')
ax.axis('off');
```

## Exploring the probabilistic nature of the sampling

```python
def generate_new_sample(z_mean : list, z_log_sigma : list):
    
    z_mean = torch.tensor(z_mean).unsqueeze(0).to(device)
    z_log_sigma = torch.tensor(z_log_sigma).unsqueeze(0).to(device)
    
    with torch.no_grad():
        z_sampled = model.sampling(z_mean, z_log_sigma)
    
        z_decoded = model.decoder(z_sampled)
    
    return z_decoded[0].cpu().numpy()
    
```

```python
a = [4, -2]
b = [0, 0]


plt.style.use('default')
fig, ax = plt.subplots(ncols = 2)

img1 = generate_new_sample(a,b)
img2 = generate_new_sample(a,b)

ax[0].imshow(img1[0], cmap = 'gray')
ax[0].axis('off')
ax[1].imshow(img2[0], cmap = 'gray')
ax[1].axis('off');
```
